```python
import Augmentor

pipeline = Augmentor.Pipeline("Test_Images")

```

```python
pipeline.random_distortion( probability =1, grid_width =4, grid_height =4, magnitude =5)
pipeline.rotate ( probability =0.5 , max_left_rotation =10 , max_right_rotation =10)
pipeline.sample(10)
```
